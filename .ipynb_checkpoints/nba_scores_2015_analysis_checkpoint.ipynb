```python
# Created by: Anthony ElHabr
# Purpose: Extract NBA teams from espn.go.com

import pandas as pd
import numpy as np
from datetime import datetime, date

YEAR = 2015
# scores_df = pd.read_csv("nba-scores-2015.csv", nrows = 6*82)
scores_df = pd.read_csv("nba-scores-2015.csv")
win_totals_df = pd.read_csv("nba-win-totals-2015.csv", index_col=False)
```

```python
default_pick = []
tony_pick = []

default_wins_to_date = []
default_losses_to_date = []
tony_wins_to_date = []
tony_losses_to_date = []

actual_sum_win = []
actual_sum_loss = []
default_sum_win = []
default_sum_loss = []
tony_sum_win = []
tony_sum_loss = []

schedule_team_tony_mu = []
schedule_team_tony_sigma = []
schedule_team_tony_win_total_pick = []
other_team_tony_mu = []
other_team_tony_sigma = []
other_team_tony_win_total_pick = []

# alternate method: all_teams = scores_df['team_schedule'].unique()
all_teams = win_totals_df['team_abbrv']

for t in all_teams:
    
    # change variable name for clarity
    this_team = t
    this_team_win_total = win_totals_df.loc\
    [win_totals_df['team_abbrv'] == this_team]['win_total']
    # need to convert type from pandas series objects to float
    this_team_win_total = float(this_team_win_total)
    
    # alternatively, could sum once at the end of for loop instead
    default_sum_win_num = 0
    default_sum_loss_num = 0
    tony_sum_win_num = 0
    tony_sum_loss_num = 0
    
    team_df = scores_df.loc[scores_df['team_schedule'] == this_team]
    
    # type of "row" would tuple without "index"
    # type of "row" is pandas series with "index"
    for index, row in team_df.iterrows():
        home_flag = row.loc['home_flag']

        if home_flag == 1:
            other_team = row.loc['away_team']
        else:
            other_team = row.loc['home_team']

        other_team_win_total = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == other_team]['win_total']

        # need to convert type from pandas series objects to float
        other_team_win_total = float(other_team_win_total)
        
        if this_team_win_total > other_team_win_total or\
        (this_team_win_total == other_team_win_total and home_flag):
            default_pick.append(this_team)
            default_wins_to_date.append('1')
            default_losses_to_date.append('0')
            default_sum_win_num += 1
        else:
            default_pick.append(other_team)
            default_wins_to_date.append('0')
            default_losses_to_date.append('1')
            default_sum_loss_num += 1
    
        this_team_tony_win_total_pick_flag = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == this_team]['tony_win_total_pick_flag']
        
        this_team_tony_win_total_pick_num = float(win_totals_df.loc\
        [win_totals_df['team_abbrv'] == this_team]['tony_win_total_pick_num'])
        
        # mean of normal distribution
        this_team_tony_mu_num = 0
        if not this_team_tony_win_total_pick_num > 0:
            if (this_team_tony_win_total_pick_flag == 'o').bool():
                this_team_tony_mu_num = int(round(this_team_win_total+1))
            else:
                this_team_tony_mu_num = int(round(this_team_win_total))
        else:
            this_team_tony_mu_num = this_team_tony_win_total_pick_num
        schedule_team_tony_mu.append(this_team_tony_mu_num)

        # standard deviation of normal distribution
        this_team_tony_sigma_num = 0
        this_team_tony_win_total_confidence = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == this_team]\
        ['tony_win_total_confidence']
        # scale tony_win_total_confidence by a reasonable factor
        this_team_tony_sigma_num =\
        float(((len(all_teams)+1)-this_team_tony_win_total_confidence)/1)
        schedule_team_tony_sigma.append(this_team_tony_sigma_num)

        # calculate normal distribution N(mu, sigma^2)            
        this_team_tony_win_total_pick_num =\
        this_team_tony_sigma_num * np.random.randn()\
        + this_team_tony_mu_num
        
        schedule_team_tony_win_total_pick.append\
        (this_team_tony_win_total_pick_num)
            
        other_team_tony_win_total_pick_flag = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == other_team]['tony_win_total_pick_flag']
        
        other_team_tony_win_total_pick_num = float(win_totals_df.loc\
        [win_totals_df['team_abbrv'] == other_team]['tony_win_total_pick_num'])
        
        # calculate mean of normal distribution
        other_team_tony_mu_num = 0        
        if not other_team_tony_win_total_pick_num > 0:
            if (other_team_tony_win_total_pick_flag == 'o').bool():
                other_team_tony_mu_num = int(round(other_team_win_total+1))
            else:
                other_team_tony_mu_num = int(round(other_team_win_total))
        else:
            other_team_tony_mu_num = other_team_tony_win_total_pick_num
        other_team_tony_mu.append(other_team_tony_mu_num)

        # calculate standard deviation of normal distributio
        other_team_tony_sigma_num = 0
        other_team_tony_win_total_confidence = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == other_team]\
        ['tony_win_total_confidence']
        # scale tony_win_total_confidence by a reasonable factor
        other_team_tony_sigma_num =\
        float(((len(all_teams)+1)-other_team_tony_win_total_confidence)/1)
        other_team_tony_sigma.append(other_team_tony_sigma_num)

        # calculate normal distribution N(mu, sigma^2)            
        other_team_tony_win_total_pick_num =\
        other_team_tony_sigma_num * np.random.randn()\
        + other_team_tony_mu_num
        
        other_team_tony_win_total_pick.append\
        (other_team_tony_win_total_pick_num)
        
        if this_team_tony_win_total_pick_num >\
        other_team_tony_win_total_pick_num or\
        (this_team_tony_win_total_pick_num ==
         other_team_tony_win_total_pick_num and home_flag):
            tony_pick.append(this_team)
            tony_wins_to_date.append('1')
            tony_losses_to_date.append('0')
            tony_sum_win_num += 1
        else:
            tony_pick.append(other_team)
            tony_wins_to_date.append('0')
            tony_losses_to_date.append('1')
            tony_sum_loss_num += 1

    actual_sum_win_num = team_df['wins_to_date'].iloc[-1]
    actual_sum_loss_num = team_df['losses_to_date'].iloc[-1]
    
    # print this_team
    # print actual_sum_win_num, default_sum_win_num, tony_sum_win_num
    # print actual_sum_loss_num, default_sum_loss_num, tony_sum_loss_num

    actual_sum_win.append(actual_sum_win_num)
    actual_sum_loss.append(actual_sum_loss_num)
    default_sum_win.append(default_sum_win_num)
    default_sum_loss.append(default_sum_loss_num)
    tony_sum_win.append(tony_sum_win_num)
    tony_sum_loss.append(tony_sum_loss_num)

print "Finished running analysis"
```
