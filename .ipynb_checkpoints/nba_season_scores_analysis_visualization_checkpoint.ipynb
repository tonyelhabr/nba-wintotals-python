```python
# Created by: Anthony ElHabr
# Purpose:

import pandas as pd
import numpy as np
# "magic" function to plot in jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

YEAR = 2015
f1 = "nba-season-scores-analysis-{0}.csv".format(YEAR)
f2 = "nba-season-scores-analysis-summary-{0}.csv".format(YEAR)
# analysis_df = pd.read_csv(f1, nrows = 6*82)
analysis_df = pd.read_csv(f1)
analysis_summary_df = pd.read_csv(f2)

analysis_df.head()
analysis_summary_df.head()
```

```python
# num_teams = analysis_df.schedule_team.unique
# num_teams = len(analysis_summary_df.schedule_team)
num_teams = len(analysis_summary_df.index)

sns.set_style("dark")

# alternative method:
# x_values = analysis_df.schedule_team.unique()
y_values = analysis_summary_df.index
x_values = analysis_summary_df.tony_sum_win

fig, ax = plt.subplots(figsize=(10,15))
title = ("Tony's Projected # of Wins (Based on Over-Under Picks)\n"\
          "for {0} - {1} Regular Season".format(YEAR-1,YEAR))
ax.set_title(title, fontsize=14, y=1.05)
ax.set_ylabel('Team', fontsize=14)
ax.set_xlabel('# of Wins', fontsize=14)
ax.set_ylim(num_teams, -1)
ax.set_xlim(0, 82)
# ax.yaxis.labelpad = 25

ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')

ax.xaxis.grid(color='white')
ax.set_axisbelow(False)

plt.yticks(range(num_teams), analysis_summary_df.schedule_team, size='large')

ax.barh(y_values, x_values, align='center')

ax.spines["top"].set_visible(False)  
ax.spines["bottom"].set_visible(False)  
ax.spines["right"].set_visible(False)  
ax.spines["left"].set_visible(False)

plt.text(0, 31, 'Created by: Anthony ElHabr', fontsize=12)
plt.show()
```
