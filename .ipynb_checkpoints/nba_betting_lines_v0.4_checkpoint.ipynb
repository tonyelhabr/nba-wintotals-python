```python
# Created by: Anthony ElHabr
# Purpose: Extract NBA betting lines from espn.go.com

# import requests
from urllib2 import urlopen
import pandas as pd
from bs4 import BeautifulSoup
import re
from datetime import datetime

SPORTSBOOKS = ['Westage', 'PinnacleSports.com', '5Dimes.eu', 'BOVADA.lv',\
               'BETONLINE.ag', 'SportsBetting.ag']

url = "http://www.espn.go.com/nba/lines"
# r = requests.get(url)
# all_data = BeautifulSoup(r.text)
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')

all_data = soup.find_all(\
                          name='tr',\
                          attrs={'class':['oddrow', 'evenrow', 'stathead']})

# all_data
```

```python
month_day = soup.find('h1').text
month_day = month_day[re.search(', ', month_day).end():]

month_day_nums = datetime.strptime(month_day, '%B %d')
matchup_date = str(month_day_nums.month) + r'/' + \
               str(month_day_nums.day) + r'/' + \
               str(datetime.now().year)
# matchup_date
```

```python
matchup_data = []
for row in range(len(all_data)): 
    for col in all_data[row].find_all('td', recursive=False):   
        matchup_data.append(col.text)
    
# matchup_data
```

```python
matchup_data_nested = []
for row in range(len(all_data)):
    extended_row = []
    
    for col in all_data[row].find_all('td', recursive=False):   
        extended_row.append(col.text)
    
    matchup_data_nested.append(extended_row)
    
# matchup_data_nested
```

```python
matchup_dict = {}
num_matchup_table_cols = 1 + len(SPORTSBOOKS)

for i in range(0, len(matchup_data_nested), num_matchup_table_cols+1):
    matchup_table = []
    for j in range(1, num_matchup_table_cols):
        matchup_table.append(matchup_data_nested[i+j])

    key = "".join(matchup_data_nested[i])
    matchup_dict[key] = matchup_table

# matchup_dict
```

```python
matchup_list_nested = []

for i in range(0, len(matchup_data_nested), num_matchup_table_cols+1):
    matchup_table = []
    for j in range(num_matchup_table_cols):
        matchup_table.append(matchup_data_nested[i+j])
    
    matchup_list_nested.append(matchup_table)

# matchup_list_nested
```

```python
import re

rotation_number = []
away_team_city = []
home_team_city = []
away_team_abbrv = []
home_team_abbrv = []
sportsbook_name = []
away_point_spread = []
home_point_spread = []
away_spread_payout = []
home_spread_payout = []
point_total = []
over_payout = []
under_payout = []
away_moneyline = []
home_moneyline = []

'''
# Example of key/value pair in matchup_dict
{u'Cleveland at Sacramento, 10:00 PM ET (514)':
 [[u'Westgate',
   u'-7+7CLE: -105SAC: -115',
   u'217o: -110u: -110',
   u'CLE: -275SAC: +225'],
  [u'PinnacleSports.com',
   u'-7+7CLE: -103SAC: -107',
   u'217o: -105u: -105',
   u'CLE: -290SAC: +256'],
  [u'5Dimes.eu',
   u'-7+7CLE: -108SAC: -102',
   u'217o: -108u: -102',
   u'CLE: -325SAC: +295'],
  [u'BOVADA.lv',
   u'-7+7CLE: -110SAC: -110',
   u'217o: 0u: 0',
   u'CLE: 0SAC: 0'],
  [u'BETONLINE.ag',
   u'-7+7CLE: -110SAC: -110',
   u'217o: -110u: -110',
   u'CLE: -300SAC: +250'],
  [u'SportsBetting.ag',
   u'-7+7CLE: -110SAC: -110',
   u'217o: -110u: -110',
   u'CLE: -300SAC: +250']],
 u'Detroit at Dallas, 8:30 PM ET (510)':
 [[u'Westgate',
   u'+2-2DET: -115DAL: -105',
   u'204.5o: -110u: -110',
   u'DET: +105DAL: -125'],
  [u'PinnacleSports.com',
   u'+2-2DET: -103DAL: -107',
  ...
   u'UTAH: 0GS: 0'],
  [u'BETONLINE.ag',
   u'+13-13UTAH: -105GS: -115',
   u'208.5o: -110u: -110',
   u'UTAH: +975GS: -1500'],
  [u'SportsBetting.ag',
   u'+13-13UTAH: -105GS: -115',
   u'208.5o: -110u: -110',
   u'UTAH: +975GS: -1500']]}
'''

for key in matchup_dict:    
    # debugging check
    if re.search('\)', key[-1]) == 'None':
        print key
    
    for i in range(len(SPORTSBOOKS)):
        # example of string to parse:
        # u'Cleveland at Sacramento, 10:00 PM ET (514)'
        rotation_number.\
        append(key[re.search('\(', key).end():-1])

        away_team_city.\
        append(key[:re.search(' at', key).start()])

        home_team_city.\
        append(key[re.search(' at ', key).end():\
                   re.search(', ', key).start()])

        # example of string to parse:
        # u'CLE: -275SAC: +225'
        away_team_abbrv.\
        append(matchup_dict[key][i][3]\
               [:\
                re.search('[A-Z]: ', matchup_dict[key][i][3]).start()+1])

        home_team_abbrv.\
        append(matchup_dict[key][i][3]\
               [re.search('[0-9][A-Z]', matchup_dict[key][i][3]).start()+1:\
                re.search('[0-9][A-Z]+:', matchup_dict[key][i][3]).end()-1])
        
        # example of string to parse:
        # u'Westgate'
        sportsbook_name.\
        append(matchup_dict[key][i][0])
        
        # example of string to parse:
        # u'-7+7CLE: -105SAC: -115'
        if matchup_dict[key][i][1] != 'EVEN':
            away_point_spread.\
            append(matchup_dict[key][i][1]\
                   [:\
                    re.search('[0-9][+-]', matchup_dict[key][i][1]).start()+1])

            home_point_spread.\
            append(matchup_dict[key][i][1]\
                   [re.search('[0-9][+-][0-9]',\
                              matchup_dict[key][i][1]).start()+1:\
                    re.search('[0-9][+-]([0-9]+)',\
                              matchup_dict[key][i][1]).end()])

            away_spread_payout.\
            append(matchup_dict[key][i][1]\
                  [re.search(': [+-]*[0-9]+[A-Z]',\
                             matchup_dict[key][i][1]).start()+2:\
                   re.search(': [+-]*[0-9]+[A-Z]',\
                             matchup_dict[key][i][1]).end()-1])

            home_spread_payout.\
            append(matchup_dict[key][i][1]\
                  [re.search(': [+-]*[0-9]+[A-Z]+: ',\
                             matchup_dict[key][i][1]).end():\
                   ])
        else:
            away_point_spread.append('0')
            home_point_spread.append('0')
            away_spread_payout.append('N/A') # moneyline payout
            home_spread_payout.append('N/A') # moneyline payout
        
        # example of string to parse:
        # u'217o: -110u: -110'
        if matchup_dict[key][i][2] != 'N/A':
            point_total.\
            append(matchup_dict[key][i][2]\
                   [:\
                    re.search('o: ', matchup_dict[key][i][2]).start()])
        
            over_payout.\
            append(matchup_dict[key][i][2]\
                  [re.search('o: ', matchup_dict[key][i][2]).end():\
                   re.search('u', matchup_dict[key][i][2]).start()])

            under_payout.\
            append(matchup_dict[key][i][2]\
                  [re.search('u: ', matchup_dict[key][i][2]).end():\
                   ])
        else:
            point_total.append('N/A')
            over_payout.append('0')
            under_payout.append('0')
        
        # example of string to parse:
        # u'CLE: -275SAC: +225'
        away_moneyline.\
        append(matchup_dict[key][i][3]\
              [re.search('[A-Z]+: ', matchup_dict[key][i][3]).end():\
               re.search('[0-9][A-Z]+:', matchup_dict[key][i][3]).start()+1])

        home_moneyline.\
        append(matchup_dict[key][i][3]\
              [re.search('[0-9][A-Z]+: ', matchup_dict[key][i][3]).end():\
               ])

'''
# for i in range(len(SPORTSBOOKS) * len(matchup_dict)):
for i in range(len(rotation_number)):
    # print away_team_city[i], home_team_city[i],\
    print matchup_date, rotation_number[i], sportsbook_name[i],\
    away_team_abbrv[i], home_team_abbrv[i],\
    away_point_spread[i], home_point_spread[i],\
    home_spread_payout[i], away_spread_payout[i],\
    point_total[i], over_payout[i], under_payout[i],\
    away_moneyline[i], home_moneyline[i]
'''

matchup_db_list = []
set_of_lists = [rotation_number, sportsbook_name,\
                away_team_abbrv, home_team_abbrv,\
                away_point_spread, home_point_spread,\
                home_spread_payout, away_spread_payout,\
                point_total, over_payout, under_payout,\
                away_moneyline, home_moneyline]


# for i in range(len(SPORTSBOOKS) * len(matchup_dict)):
for i in range(len(rotation_number)):
    single_item = []
    
    run_timestamp = datetime.now()
    run_timestamp = run_timestamp.strftime('%m/%d/%Y, %H:%M:%S')
    
    single_item.append(run_timestamp)
    single_item.append(matchup_date)
    
    for j in set_of_lists:
        single_item.append(j[i])
    
    matchup_db_list.append(single_item)

# matchup_db_list
```

```python
# TODO: consider making the matchup_db_list to a dict where the key is
# matchup_date + rotation_number + sporstbook_name because pandas DataFrames
# are more naturally constructed from dicts
df_col_headers = ['run_timestamp', 'matchup_date',\
                  'rotation_number', 'sportsbook_name',\
                  'away_team_abbrv', 'home_team_abbrv',\
                  'away_point_spread', 'home_point_spread',\
                  'home_spread_payout', 'away_spread_payout',\
                  'point_total', 'over_payout', 'under_payout',\
                  'away_moneyline', 'home_moneyline']
# df_col_headers = [header.upper() for header in df_col_headers]
matchup_df = pd.DataFrame(matchup_db_list, columns=df_col_headers)

# matchup_df
```
