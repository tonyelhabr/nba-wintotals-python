```python
# Created by: Anthony ElHabr
# Purpose:

import pandas as pd
import numpy as np
from datetime import datetime, date
import os
import matplotlib.pyplot as plt
import seaborn as sns

YEAR = 2014
f1 = "nba-season-scores-{0}.csv".format(YEAR)
f2 = "nba-season-win-totals-{0}.csv".format(YEAR)
# scores_df = pd.read_csv(filename, nrows = 6*82)
scores_df = pd.read_csv(f1)
win_totals_df = pd.read_csv(f2, index_col=False)
```

```python
start_all = datetime.now()

# TODO: do a running total model

default_pick = []
default2_pick = []
tony_pick = []

default_correct_flag = []
default2_correct_flag = []
tony_correct_flag = []

default_wins_to_date = []
default_losses_to_date = []
default2_wins_to_date = []
default2_losses_to_date = []
tony_wins_to_date = []
tony_losses_to_date = []

actual_sum_win = []
actual_sum_loss = []
default_sum_win = []
default_sum_loss = []
default_sum_win_error = []
default_win_total_correct_flag = []
default2_sum_win = []
default2_sum_loss = []
default2_sum_win_error = []
default2_win_total_correct_flag = []
tony_sum_win = []
tony_sum_loss = []
tony_sum_win_error = []
tony_win_total_correct_flag = []

schedule_team_tony_mu = []
schedule_team_tony_sigma = []
schedule_team_tony_win_total_pick = []
other_team_tony_mu = []
other_team_tony_sigma = []
other_team_tony_win_total_pick = []

# alternate method: all_teams = scores_df['team_schedule'].unique()
all_teams = win_totals_df['team_abbrv']

for t in all_teams:
    start = datetime.now()
    
    # change variable name for clarity
    this_team = t
    this_team_win_total = win_totals_df.loc\
    [win_totals_df['team_abbrv'] == this_team]['win_total']
    # need to convert type from pandas series objects to float
    this_team_win_total = float(this_team_win_total)
    
    # alternatively, could sum once at the end of for loop instead
    default_sum_win_num = 0
    default_sum_loss_num = 0
    default2_sum_win_num = 0
    default2_sum_loss_num = 0
    tony_sum_win_num = 0
    tony_sum_loss_num = 0
    
    team_df = scores_df.loc[scores_df['team_schedule'] == this_team]
    
    # type of "row" would tuple without "index"
    # type of "row" is pandas series with "index"
    for index, row in team_df.iterrows():
        home_flag = row.loc['home_flag']
        win_flag = row.loc['win_flag']

        if home_flag == 1:
            other_team = row.loc['away_team']
        else:
            other_team = row.loc['home_team']

        other_team_win_total = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == other_team]['win_total']

        # need to convert type from pandas series objects to float
        other_team_win_total = float(other_team_win_total)

        # TODO: make this into  a function
        # START FUNCTION #1 HERE
        default_correct_boolean = -1
        if this_team_win_total > other_team_win_total or\
        (this_team_win_total == other_team_win_total and home_flag):
            default_pick.append(this_team)
            default_wins_to_date.append('1')
            default_losses_to_date.append('0')
            default_sum_win_num += 1         
            if win_flag == 1:
                default_correct_boolean = 1
            elif win_flag == 0:
                default_correct_boolean = 0
            default_correct_flag.append(default_correct_boolean)
        else:
            default_pick.append(other_team)
            default_wins_to_date.append('0')
            default_losses_to_date.append('1')
            default_sum_loss_num += 1
            if win_flag == 0:
                default_correct_boolean = 1
            elif win_flag == 1:
                default_correct_boolean = 0
            default_correct_flag.append(default_correct_boolean)
        # END FUNCTION #1 HERE
        
        this_team_wins_to_date = row.loc['wins_to_date']
        game_date = row.loc['game_date']
        other_team_df = scores_df.loc[scores_df['team_schedule'] == other_team]
        other_team_wins_to_date = -1
        other_team_wins_to_date = int(other_team_df.\
        loc[other_team_df['game_date'] == game_date]['wins_to_date'])

        # CALL FUNCTION #1 HERE
        default2_correct_boolean = -1
        if this_team_wins_to_date > other_team_wins_to_date or\
        (this_team_wins_to_date == other_team_wins_to_date and home_flag):
            default2_pick.append(this_team)
            default2_wins_to_date.append('1')
            default2_losses_to_date.append('0')
            default2_sum_win_num += 1         
            if win_flag == 1:
                default2_correct_boolean = 1
            elif win_flag == 0:
                default2_correct_boolean = 0
            default2_correct_flag.append(default2_correct_boolean)
        else:
            default2_pick.append(other_team)
            default2_wins_to_date.append('0')
            default2_losses_to_date.append('1')
            default2_sum_loss_num += 1
            if win_flag == 0:
                default2_correct_boolean = 1
            elif win_flag == 1:
                default2_correct_boolean = 0
            default2_correct_flag.append(default2_correct_boolean)
        # END FUNCTION #1 HERE
            
        # TODO: make this into a function
        # START FUNCTION #2 HERE
        this_team_tony_win_total_pick_flag = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == this_team]['tony_win_total_pick_flag']
        
        # this_team_tony_win_total_pick_num = float(win_totals_df.loc\
        # [win_totals_df['team_abbrv'] == this_team]['tony_win_total_pick_num'])
        
        # mean of normal distribution
        this_team_tony_mu_num = 0
        if (this_team_tony_win_total_pick_flag == 'o').bool():
            this_team_tony_mu_num = int(round(this_team_win_total+1))
        else:
            this_team_tony_mu_num = int(round(this_team_win_total))
        
        # account for home court advantage
        if home_flag == 1:
            this_team_tony_mu_num += 2
        elif home_flag == 0:
            this_team_tony_mu_num -= 2
        schedule_team_tony_mu.append(this_team_tony_mu_num)

        # standard deviation of normal distribution
        this_team_tony_sigma_num = 0
        this_team_tony_win_total_confidence = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == this_team]\
        ['tony_win_total_confidence']

        # account for 'tony_win_total_confidence' = NaN
        # in older 'nba-season-win-totals-yyyy.csv' files
        if this_team_tony_win_total_confidence.empty:
            this_team_tony_win_total_confidence = len(all_teams)/2
            print "No win_total_confidence"
        
        # scale tony_win_total_confidence by a reasonable factor
        this_team_tony_sigma_num =\
        float(((len(all_teams)+1)-this_team_tony_win_total_confidence)/1)
        schedule_team_tony_sigma.append(this_team_tony_sigma_num)

        # calculate normal distribution N(mu, sigma^2)            
        this_team_tony_win_total_pick_num =\
        this_team_tony_sigma_num * np.random.randn()\
        + this_team_tony_mu_num
        # END FUNCTION #2 HERE
        
        # CALL FUNCTION #2 HERE
        schedule_team_tony_win_total_pick.append\
        (this_team_tony_win_total_pick_num)
            
        other_team_tony_win_total_pick_flag = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == other_team]['tony_win_total_pick_flag']
        
        # other_team_tony_win_total_pick_num = float(win_totals_df.loc\
        # [win_totals_df['team_abbrv'] == other_team]['tony_win_total_pick_num'])
        
        # calculate mean of normal distribution
        other_team_tony_mu_num = 0        
        if (other_team_tony_win_total_pick_flag == 'o').bool():
            other_team_tony_mu_num = int(round(other_team_win_total+1))
        else:
            other_team_tony_mu_num = int(round(other_team_win_total))
            
        # account for home court advantage
        if home_flag == 1:
            other_team_tony_mu_num -= 2
        elif home_flag == 0:
            other_team_tony_mu_num += 2
        other_team_tony_mu.append(other_team_tony_mu_num)

        # calculate standard deviation of normal distribution
        other_team_tony_sigma_num = 0
        other_team_tony_win_total_confidence = win_totals_df.loc\
        [win_totals_df['team_abbrv'] == other_team]\
        ['tony_win_total_confidence']

        # account for 'tony_win_total_confidence' = NaN
        # in older 'nba-season-win-totals-yyyy.csv' files
        if other_team_tony_win_total_confidence.empty:
            other_team_tony_win_total_confidence = len(all_teams)/2
            print "No win_total_confidence"

        # scale tony_win_total_confidence by a reasonable factor
        other_team_tony_sigma_num =\
        float(((len(all_teams)+1)-other_team_tony_win_total_confidence)/1)
        other_team_tony_sigma.append(other_team_tony_sigma_num)

        # calculate normal distribution N(mu, sigma^2)            
        other_team_tony_win_total_pick_num =\
        other_team_tony_sigma_num * np.random.randn()\
        + other_team_tony_mu_num
        
        other_team_tony_win_total_pick.append\
        (other_team_tony_win_total_pick_num)
        # END FUNCTION HERE #2

        # CALL FUNCTION #1 HERE
        tony_correct_boolean = -1
        if this_team_tony_win_total_pick_num >\
        other_team_tony_win_total_pick_num or\
        (this_team_tony_win_total_pick_num ==
         other_team_tony_win_total_pick_num and home_flag):
            tony_pick.append(this_team)
            tony_wins_to_date.append('1')
            tony_losses_to_date.append('0')
            tony_sum_win_num += 1
            if win_flag == 1:
                tony_correct_boolean = 1
            elif win_flag == 0:
                tony_correct_boolean = 0
            tony_correct_flag.append(tony_correct_boolean)
        else:
            tony_pick.append(other_team)
            tony_wins_to_date.append('0')
            tony_losses_to_date.append('1')
            tony_sum_loss_num += 1
            if win_flag == 1:
                tony_correct_boolean = 0
            elif win_flag == 0:
                tony_correct_boolean = 1
            tony_correct_flag.append(tony_correct_boolean)
        # END FUNCTION #1 HERE

    actual_sum_win_num = team_df['wins_to_date'].iloc[-1]
    actual_sum_loss_num = team_df['losses_to_date'].iloc[-1]
    
    default_sum_win_error_num = abs(actual_sum_win_num-default_sum_win_num)
    default2_sum_win_error_num = abs(actual_sum_win_num-default2_sum_win_num)
    tony_sum_win_error_num = abs(actual_sum_win_num-tony_sum_win_num)

    # TODO: make this into a function
    # START FUNCTION #3 HERE
    # boolean == -1 indicates a tie 
    default_win_total_correct_boolean = -1
    if actual_sum_win_num > this_team_win_total:
        if default_sum_win_num > this_team_win_total:
            default_win_total_correct_boolean = 1
        elif default_sum_win_num < this_team_win_total:
            default_win_total_correct_boolean = 0
    elif actual_sum_win_num < this_team_win_total:
        if default_sum_win_num < this_team_win_total:
            default_win_total_correct_boolean = 1
        elif default_sum_win_num > this_team_win_total:
            default_win_total_correct_boolean = 0
    # END FUNCTION #3 HERE

    # CALL FUNCTION #3 HERE
    default2_win_total_correct_boolean = -1
    if actual_sum_win_num > this_team_win_total:
        if default2_sum_win_num > this_team_win_total:
            default2_win_total_correct_boolean = 1
        elif default2_sum_win_num < this_team_win_total:
            default2_win_total_correct_boolean = 0
    elif actual_sum_win_num < this_team_win_total:
        if default2_sum_win_num < this_team_win_total:
            default2_win_total_correct_boolean = 1
        elif default2_sum_win_num > this_team_win_total:
            default2_win_total_correct_boolean = 0
    # END FUNCTION #3 HERE
    
    # CALL FUNCTION #3 HERE
    tony_win_total_correct_boolean = -1
    if actual_sum_win_num > this_team_win_total:
        if tony_sum_win_num > this_team_win_total:
            tony_win_total_correct_boolean = 1
        elif tony_sum_win_num < this_team_win_total:
            tony_win_total_correct_boolean = 0
    elif actual_sum_win_num < this_team_win_total:
        if tony_sum_win_num < this_team_win_total:
            tony_win_total_correct_boolean = 1
        elif tony_sum_win_num > this_team_win_total:
            tony_win_total_correct_boolean = 0
    # END FUNCTION #3 HERE
    
    # print this_team
    # print actual_sum_win_num, default_sum_win_num, tony_sum_win_num
    # print actual_sum_loss_num, default_sum_loss_num, tony_sum_loss_num

    actual_sum_win.append(actual_sum_win_num)
    actual_sum_loss.append(actual_sum_loss_num)
    default_sum_win.append(default_sum_win_num)
    default_sum_loss.append(default_sum_loss_num)
    default_sum_win_error.append(default_sum_win_error_num)
    default_win_total_correct_flag.append(default_win_total_correct_boolean)
    default2_sum_win.append(default2_sum_win_num)
    default2_sum_loss.append(default2_sum_loss_num)
    default2_sum_win_error.append(default2_sum_win_error_num)
    default2_win_total_correct_flag.append(default2_win_total_correct_boolean)
    tony_sum_win.append(tony_sum_win_num)
    tony_sum_loss.append(tony_sum_loss_num)
    tony_sum_win_error.append(tony_sum_win_error_num)
    tony_correct_flag.append(tony_correct_boolean)
    tony_win_total_correct_flag.append(tony_win_total_correct_boolean)
    
    end = datetime.now()
    time_diff = end - start
    print "Finished running analysis for {0} in {1} s".format(t, time_diff)

end_all = datetime.now()
time_diff_all = end_all - start_all
print "Finished running all analysis in {0} s".format(time_diff_all)
```

```python
df_col_headers = ['schedule_team',\
                  'actual_sum_win', 'default_sum_win',\
                  'default2_sum_win', 'tony_sum_win',\
                  'actual_sum_loss', 'default_sum_loss',\
                  'default2_sum_loss', 'tony_sum_loss',\
                  'default_sum_win_error',\
                  'default2_sum_win_error',\
                  'tony_sum_win_error',\
                  'default_win_total_correct_flag',\
                  'default2_win_total_correct_flag',\
                  'tony_win_total_correct_flag',\
                  'schedule_team_tony_win_total_pick_mean',\
                  'schedule_team_tony_mu_mean', 'schedule_team_tony_sigma_mean',\
                  'other_team_tony_win_total_pick_mean',\
                  'other_team_tony_mu_mean', 'other_team_tony_sigma_mean']

analysis_summary_df = pd.DataFrame\
(analysis_summary_db_list, columns=df_col_headers)

sum_row = analysis_summary_df[df_col_headers[1:]].sum()
mean_row= analysis_summary_df[df_col_headers[1:]].mean()

sum_df = pd.DataFrame(sum_row).T
sum_df = sum_df.reindex(columns=analysis_summary_df.columns)
mean_df = pd.DataFrame(mean_row).T
mean_df = mean_df.reindex(columns=analysis_summary_df.columns)

analysis_summary_df2 = analysis_summary_df.append(sum_df, ignore_index=True)
analysis_summary_df3 = analysis_summary_df2.append(mean_df, ignore_index=True)

analysis_summary_df3
```
