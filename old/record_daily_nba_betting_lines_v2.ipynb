```python
# Created by: Anthony ElHabr
# Purpose: Extract NBA betting lines from espn.go.com

from urllib2 import urlopen
import pandas as pd
from bs4 import BeautifulSoup
import re

TEAMS = ['ATL', 'BOS', 'BKN', 'CHA', 'CHI', 'CLE', 'DAL', 'DEN',\
        'DET', 'GSW', 'HOU', 'IND', 'LAC', 'LAL', 'MEM', 'MIA',\
        'MIL', 'MIN', 'NOP', 'NYK', 'OKC', 'ORL', 'PHI', 'PHX',\
        'POR', 'SAC', 'SAS', 'TOR', 'UTA', 'WAS']
BOOKS = ['Westage', 'PinnacleSports.com', '5Dimes.eu', 'BOVADA.lv'\
        'BETONLINE.ag', 'SportsBetting.ag']

url = "http://www.espn.go.com/nba/lines"
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')

html_v2 = urlopen(url)
html_v2 = html_v2.read()
html_v2 = re.sub('<br/>', '</td><td>', html_v2)
soup_v2 = BeautifulSoup(html_v2, 'html.parser')
```

```python
matchup_data = []

for row in range(len(all_rows)):
    matchup_rows = []
    
    for cell in all_rows[row].find_all('td'):
        matchup_rows.append(cell.text)
    
    matchup_data.append(matchup_rows)

matchup_data
```
