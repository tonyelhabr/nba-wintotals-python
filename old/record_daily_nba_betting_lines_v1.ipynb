```python
url = "http://www.espn.go.com/nba/lines"
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
```

```python
main_headers = []

all_headers = soup.find_all(class_="colhead")[1].findAll('td')
for td in all_headers:
    main_headers.append(td.text)

main_headers
```

```python
# list comprehension to get column headers
main_headers_2 = [td.get_text() for td in \
               soup.find_all(class_="colhead")[1].findAll('td')]

if main_headers != main_headers_2:
    print main_headers_2
```

```python
# THIS METHOD GIVES THE DESIRED OUTPUT WITH EXTRA DATA
all_rows = soup.find_all('tr')
# create an empty list for all matchups
matchup_data = []

# search through each table row
for row in range(len(all_rows)):
    # create an empty list for each matchup
    matchup_rows = []
    
    # for each table data elment from each table row
    for td in all_rows[row].find_all('td'):
        # get the text content and append to the matchup_rows list
        matchup_rows.append(td.text)
    
    # then append each matchup to the matchup_data list
    matchup_data.append(matchup_rows)

# matchup_data
```

```python
# THIS METHOD GIVES THE DESIRED OUTPUT WITH EXTRA DATA
# list comprehension to get matchup data
matchup_data_2 = [[td.text for td in all_rows[row].find_all('td')]\
             for row in range(len(all_rows))]

if matchup_data != matchup_data_2:
    print matchup_data_2
```
